```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('car.csv')
df
```

```python
df.columns
```

```python
columns=['price','maintainence','doors','person','luggage','safety','decision']
df.columns=columns
```

```python
df
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
for i in df.columns:
    df[i]=label_encoder.fit_transform(df[i])
df
```

```python
for i in df.columns:
    print(i,' ',df[i].unique())
```

```python
x=df.drop('decision',axis=1).values
y=df['decision']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=.7,test_size=.3)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
classifier=GradientBoostingClassifier()
classifier.fit(x_train,y_train)
```

```python
trained_values=classifier.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score
print("accuracy is ",accuracy_score(y_train,trained_values))
```

```python
print("accuracy on test data is ",accuracy_score(y))
```
